```python
import cv2
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
kernel_size = 5
gauss_img = cv2.GaussianBlur(gray_img,(kernel_size, kernel_size), 0)
```

```python
low_threshold, high_threshold = [200, 300]
canny_img = cv2.Canny(gauss_img, low_threshold, high_threshold)
print(canny_img.shape)
```

```python
plt.imshow(canny_img, cmap='gray')
plt.show()
```

```python
import numpy as np
vertices = np.array([[(0, canny_img.shape[0]), (canny_img.shape[1], canny_img.shape[0]), (400, 260), (600, 260)]])
```

```python
mask = np.zeros_like(canny_img)
```

```python
mask_channels = (255,) * img.shape[2]
```

```python
a = cv2.fillPoly(mask, vertices, mask_channels)
```

```python
plt.imshow(a, cmap='gray')
plt.show()
```

```python
print(a.shape)
print(canny_img.shape)
```

```python
masked_img = cv2.bitwise_and(canny_img, a)
```

```python
plt.imshow(masked_img, cmap='gray')
plt.show()
```
